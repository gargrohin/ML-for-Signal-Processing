```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.linalg import eigh
```

```python
df = pd.read_csv('mnist_train.csv')
```

```python
train = []
for i in range(10):
    train.append([])
```

```python
test = []
for i in range(10):
    test.append([])
```

```python
for index, row in df.iterrows():
        l = row[0]
        a = row[1:]
        train[l].append(np.array(a).reshape(28*28))
```

```python
pic = train[4][10].reshape(28,28) # reshape from 1d to 2d pixel array
```

```python
plt.figure()
plt.imshow(pic)
```

```python
mat = []
for x in range(10):
    for a in train[x]:
        mat.append(a)
```

```python
mat = np.array(mat)
```

```python
mean = np.mean(mat, axis = 0)
mat = mat - mean
```

```python
sig = (np.std(mat, axis= 0) + 10**(-10))
```

```python
mat = mat / sig
```

```python
sigma = np.dot(mat.T,mat)
lamda , vec = eigh(sigma, eigvals=(782,783))
```

```python
reduced = np.dot(mat,vec)
```

```python
reduced.shape
```

```python
# transformed data
```

```python
fig = plt.figure()
k= 0
for i in range(10):
    plt.scatter(reduced[k:k+len(train[i])-1,0], reduced[k:k+len(train[i])-1,1], label = i, s = 5)
    k = k + len(train[i])
fig.legend()
plt.show()
```

```python
img = vec[:,0].reshape(28,28)
plt.imshow(img)
plt.title('bases 1')
plt.show()
img = vec[:,1].reshape(28,28)
plt.imshow(img)
plt.title('bases 2')
plt.show()
```

```python
recons = np.dot(reduced,vec.T)
recons = recons*(sig)
recons+=mean
```

```python
k = 0
fig=plt.figure(figsize=(10, 10))
for i in range(10):
    fig.add_subplot(1,10,i+1)
    img = recons[k].reshape(28,28)
    k = k + len(train[i])
    plt.imshow(img)
    plt.title(str(i))
    plt.axis('off')
    
```
