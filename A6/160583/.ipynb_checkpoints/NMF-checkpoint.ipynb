```python
def euclid(p, q):
    return np.linalg.norm(p-q)
```

```python
def nmf_fit(V,W,H,numberOfIterations,recoError,eps):
    for n in np.arange(numberOfIterations):

#         print("iteration ", n, " over ", numberOfIterations)


        temp_H = np.dot(W.T , V)
        H_den = np.maximum(np.dot(W.T , W) , eps)
        H_den = np.maximum(np.dot(H_den , H), eps)

        H = H*(temp_H / H_den)

        temp_W = np.dot(V, H.T)
        W_den = np.maximum(np.dot(W , H) , eps)
        W_den = np.maximum(np.dot(W_den , H.T), eps)

        W = W*(temp_W / W_den)

        hatV = np.dot(W,H)

        recoError.append(euclid(V , hatV))

#         print("Reconstruction error  : ", recoError[n])
        
    return W, H, hatV, recoError
```

```python
np.random.seed(0)
# bases = []
weights_train = []
reco_error = []
nmf_components = 2
# for i in range(10):
# print('label)
numberOfIterations = 50
# train_samples = len(train[i])
V = []
for x in train:
    for y in x:
        V.append(y)
V = np.array(V)
V = V.T
W = np.random.rand(V.shape[0] , nmf_components)
H = np.random.rand(nmf_components, V.shape[1])
recoError = []
eps = 10 ** (-20)

W, H, _, recoError = nmf_fit(V, W, H, numberOfIterations, recoError, eps)

# bases.append(W)
weights_train.append(H)
# reco_error.append(recoError)
    
    
    
```

PART B

classification on test data

```python
np.random.seed(0)
bases = []
weights_train = []
reco_error = []
nmf_components = 5
for i in range(10):
#     print('label')
    numberOfIterations = 50
    # train_samples = len(train[i])
    V = []
    for x in train[i]:
      V.append(x)
    V = np.array(V)
    V = V.T
    W = np.random.rand(V.shape[0] , nmf_components)
    H = np.random.rand(nmf_components, V.shape[1])
    recoError = []
    eps = 10 ** (-20)

    W, H, _, recoError = nmf_fit(V, W, H, numberOfIterations, recoError, eps)

    bases.append(W)
    weights_train.append(H)
    reco_error.append(recoError)
    
    
    
```

```python
def nmf_predict(V,W,H,numberOfIterations,recoError,eps):
    for n in np.arange(numberOfIterations):

#         print("iteration ", n, " over ", numberOfIterations)


        temp_H = np.dot(W.T , V)
        H_den = np.maximum(np.dot(W.T , W) , eps)
        H_den = np.maximum(np.dot(H_den , H), eps)

        H = H*(temp_H / H_den)

#         temp_W = np.dot(V, H.T)
#         W_den = np.maximum(np.dot(W , H) , eps)
#         W_den = np.maximum(np.dot(W_den , H.T), eps)

#         W = W*(temp_W / W_den)

        hatV = np.dot(W,H)

        recoError.append(euclid(V , hatV))

#         print("Reconstruction error  : ", recoError[n])
        
    return W, H, hatV, recoError
```

```python
def predict(H):
    
    ans = []
    for k in range(H.shape[1]):
        ans.append(0)
        max_far = 0
        j = 0
        for i in range(10):
            h = H[j:j+nmf_components,k]
            j = j+nmf_components
            if np.linalg.norm(h)>max_far:
                max_far = np.linalg.norm(h)
                ans[k] = i
    return ans
```

```python
np.random.seed(0)
pred = []
for i in range(10):
    pred.append([])
    
W = np.concatenate((bases[0], bases[1]), 1)
for i in range(2,10):
    W = np.concatenate((W,bases[i]),1)

for l in range(10):
    print("test label: " , l)
    samples = np.array(test[l]).T
    H = np.random.rand(nmf_components*10, samples.shape[1])
    recoError = []
    eps = 10 ** (-20)
    
    _, H, _,_ = nmf_predict(samples,W,H,50,recoError,eps)
    
    labels = predict(H)
    pred[l] = labels
    
    
```

```python
tot_acc = 0
total = 0
for i in range(10):
    acc = 0
    total += len(pred[i])
    for x in pred[i]:
        if x == i:
            acc+=1
    tot_acc+=acc
    
    print('accuracy for label: ' , i ,': ', acc/len(pred[i]))

print('total accuracy: ', tot_acc/total)
```

Thus, total accuracy = 88.26%
